```python
from datasets import load_dataset
file_path = '/content/drive/My Drive/Hackathon_Project/training.csv'
dataset = load_dataset('csv', data_files=file_path)
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('gpt2')
tokenizer.pad_token = tokenizer.eos_token
```

```python
def preprocess_function(examples):
    text_inputs = [prompt + completion + tokenizer.eos_token for prompt, completion in zip(examples['prompt'], examples['completion'])]
    result = tokenizer(text_inputs, truncation=True, padding='max_length', max_length=128)
    result["labels"] = result["input_ids"].copy()
    return result

tokenized_dataset = dataset.map(preprocess_function, batched=True, remove_columns=dataset['train'].column_names)
```

```python
from transformers import AutoModelForCausalLM
model = AutoModelForCausalLM.from_pretrained('gpt2')
```

```python
from transformers import TrainingArguments
training_args = TrainingArguments(
    output_dir='/content/drive/My Drive/Hackathon_Project/results_new',
    num_train_epochs=8,
    per_device_train_batch_size=2,
    warmup_steps=50,
    weight_decay=0.01,
    logging_dir='./logs',
    report_to="none",
)
```

```python
from transformers import Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset['train'],
)

trainer.train()
```

```python
save_path = '/content/drive/My Drive/Hackathon_Project/my-food-chatbot-model-v3'
trainer.save_model(save_path)
tokenizer.save_pretrained(save_path)
print("New model successfully saved!")
```
